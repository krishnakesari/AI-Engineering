```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.1)
# adjusting slope and intercept to verify changes in the graph
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size = x.size)
ydata = y + y_noise
# plotting figures (figsize = (8,6))
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
# Cubic function y = ax3 + bx2 + cx + d
x = np.arange(-5, 5, 1)
# adjusting slope and intercept to verify changes in the graph
y = 1 * (x ** 3) + 1 * (x ** 2) + 1 * x + 3
y_noise = 20 * np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
# Quadratic function Y = X2 
x = np.arange(-5, 5, 0.1)
y = np.power(x,2)
y_noise = 2 * np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
# Exponential Y = a + bc^X
X = np.arange(-5,5,0.1)
Y = np.exp(X)
plt.plot(X, Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
# Logarithemic Y = log(X)
X = np.arange(-5, 5, 0.1)
Y = np.log(X)
plt.plot(X, Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent variable')
plt.show()
```

```python
# Sigmoidal/Logistic Y = a + b / (1 + c^(X-d))
X = np.arange(-5, 5, 0.1)
Y = 1 - 4/(1 + np.power(3, X - 2))
plt.plot(X, Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
# As the above plot looks like logitistic regression
X = np.arange(-5, 5, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))

plt.plot(X, Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```
